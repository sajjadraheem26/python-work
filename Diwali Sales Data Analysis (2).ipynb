```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

```

```python
df=pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
#drop coplumn 13 and 14 having 0 values
df.drop(['Status','unnamed1'], axis=1,inplace=True)
```

```python
df.shape
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
# thus amount column has 12 null values, thus it need to be droped.
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# change datatype of amount column from float to whole no.
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.describe()
# It gives mathematical idea of numerical column
```

```python
df[['Age', 'Orders', 'Amount']].describe()
```

# EDA


```python
plot_of_gender= sb.countplot(x='Gender' ,data=df)

for i in plot_of_gender.containers:
    plot_of_gender.bar_label(i)
```

```python
#The count of female buyers are more than twice than male.
```

```python
sorted_values=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sorted_values
```

```python
# The amount spent by female is also more than twice than of male.
```

```python
sb.barplot(x='Gender',y='Amount', data=sorted_values)

```

```python
# most of the buyer are female either purchasing by amount or by count
```

```python
order1=['0-17','18-25','26-35','36-45','46-50','51-55','55+']
plot_by_age=sb.countplot(x='Age Group',data=df,hue='Gender', order=order1)
for i in plot_by_age.containers:
    plot_by_age.bar_label(i)
    
```

```python
# data of amount vs age group
sorted_values_AG=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sorted_values_AG
```

```python
order_2=['0-17','18-25','26-35','36-45','46-50','51-55','55+']
sb.barplot(x='Age Group',y='Amount' ,data=sorted_values_AG ,order=order_2)
```

```python
# most of the buyers are b/w age of 26-35
```

```python
#amount state wise 
sorted_values_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sorted_values_state
```

```python
sb.barplot(y='State', x='Amount', data=sorted_values_state)
```

```python
# order statewise
sorted_order_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sorted_order_state
```

```python
sb.barplot(x='Orders', y='State', data=sorted_order_state)
```

```python
# so uttar pradesh on top based on no of orders and amount spent.
```

```python
#orders & amount based on marital status 
#0 for married
sorted_order_ms=df.groupby(['Marital_Status'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sorted_order_ms
```

```python
sb.barplot(x='Marital_Status', y='Orders', data=sorted_order_ms)
```

```python
sorted_amount_ms=df.groupby(['Marital_Status'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sorted_amount_ms
```

```python
sb.barplot(x='Marital_Status', y='Amount', data=sorted_amount_ms)
plt.show()
```

```python
plot_of_count_on_ms = sb.countplot(x='Marital_Status', data=df, hue='Gender')
for i in plot_of_count_on_ms.containers:
    plot_of_count_on_ms.bar_label(i)
```

```python
#married women are top in terms or no of order or by amount too
```

```python
orders_on_occup=sb.countplot(y='Occupation', data=df)
for i in orders_on_occup.containers:
    orders_on_occup.bar_label(i)
```

```python
sorted_amount_occup=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sorted_amount_occup
```

```python
sb.barplot(y='Occupation',x='Amount', data=sorted_amount_occup)
```

```python
orders_on_occup=sb.countplot(y='Occupation', data=sorted_amount_occup)
for i in orders_on_occup.containers:
    orders_on_occup.bar_label(i)
```

```python
#It sector on top folowed by healthcare in termns of no of orders and amount
```

```python
orders_on_category=sb.countplot(y='Product_Category', data=df)
for i in orders_on_category.containers:
    orders_on_category.bar_label(i)
```
